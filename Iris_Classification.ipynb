#Import libraries

```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
#Load the iris dataset from sklearn
iris=datasets.load_iris()
```

```python
#To check the features of the dataset
iris.feature_names
```

```python
inputs=iris.data
inputs
```

```python
#To check the target names
iris.target_names
```

```python
outputs=iris.target
outputs
```

```python
inputs.shape
```

```python
outputs.shape
```

#Train_Test_Split

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(inputs,outputs,test_size=0.2,random_state=1)
```

```python
Xtrain[0]
```

```python
ytrain[0]
```

#Neural Network Training

```python
network=MLPClassifier(max_iter=2000,activation="logistic",solver="adam",learning_rate="constant",learning_rate_init=0.001,hidden_layer_sizes=(4,4),batch_size=30,verbose=True)
network.fit(Xtrain,ytrain)
```

#Visualizing the reduction in loss/iteration

```python
plt.plot(network.loss_curve_)
plt.title("Loss Curve for MLP Classifier")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
#To check the activation function used at the output layer
network.out_activation_
```

#Model Evaluation

```python
predictions=network.predict(Xtest)
predictions
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score of the model", accuracy_score(ytest,predictions))
```

```python
from yellowbrick.classifier import ConfusionMatrix
conf_mtx=ConfusionMatrix(network,classes=iris.target_names)
conf_mtx.fit(Xtrain,ytrain)
conf_mtx.score(Xtest,ytest)
conf_mtx.show()
```
